### Data Visualization

### Count plot for Class

```python
plt.figure(dpi=100)
sns.countplot(df['Class'])
```

### Checking Class values

```python
print("No of instance of 1's : ",len(df[df['Class']==1]))
print("No of instance of 0's : ",len(df[df['Class']==0]))
```

### Bar Plot b/w Amount and class

```python
plt.figure(dpi=100)
sns.barplot(x='Class',y="Amount",data=df)
plt.legend(title='Class vs Amount')
```

```python
plt.figure(dpi=80)
sns.scatterplot(x="Time",y="Amount",data=df)
plt.figure(dpi=80)
sns.scatterplot(x="Class",y="Amount",data=df)
```
